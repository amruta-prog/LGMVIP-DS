```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.hist(Iris['SepalLengthCm'], color='red')
ax1.set_xlabel('SepalLengthCm in CM')

ax2.hist(Iris['SepalWidthCm'], color='blue')
ax2.set_xlabel('SepalWidthCm in CM')


ax3.hist(Iris['PetalLengthCm'], color='yellow')
ax3.set_xlabel('PetalLengthCm in CM')


ax4.hist(Iris['PetalWidthCm'], color='grey')
ax4.set_xlabel('PetalWidthCm in CM')

fig.set_figheight(15)

plt.show()
```

```python
corr=Iris.corr()
plt.figure(figsize = (5,6))
sns.heatmap(corr , annot = True, vmin = -0.1, cmap = "mako")
plt.title('correlation Matrix')
plt.show()
```
